```
k=4
eta=0.01
epochs=0.000001
delta_k=1
max_iters=100
iteration=0
  

 
while abs(delta_k)>epochs and iteration<max_iters:
   prev_k=k
   derivative=4*((prev_k)**3)+6*(prev_k)
   delta_k=-eta*derivative
   k=k+delta_k
   iteration=iteration+1
   print("Iteration",iteration,"\nK value is ", k)
print("the local minimum occurs at", k)
```
